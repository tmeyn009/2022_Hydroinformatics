```python
import numpy as np
from scipy.optimize import minimize
```

```python
def rosen(x, a=1.0, b=100.0):
    """The Rosenbrock function"""
    return (a - x[0]) ** 2 + b * (x[1] - x[0] ** 2) ** 2
```

```python
x0 = np.array([1.25, 0.2])
res = minimize(rosen, x0, options={'disp': True})
```

```python
print(res.x)
```
